```python
# lets get our imports
import os
import pandas as pd
import numpy as np

%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
x = [2,7,12,17,22,27,32,37,42,47,52,57,62,67,72,77,82,87,92]
y = [3.2, 3.4, 3.4, 3.3, 3.3, 3.6, 4.3, 4.5, 4.1, 3.7, 2.8, 2.3, 2.1, 2.0, 1.9, 1.4, 1.0, 0.5, 0.3]


fig = plt.figure(figsize=(10, 10), dpi= 80, facecolor='w', edgecolor='k')
plt.scatter(x, y, s = 75, c = 'black', alpha = 1, lw = 0)
plt.plot(np.unique(x), np.poly1d(np.polyfit(x, y, 2))(np.unique(x)))
plt.title("percentage of women of total female population by age group")
plt.ylabel("percent of women")
plt.xlabel("age")
plt.show()
```

```python
y = [13,23,21,27,37,52,34,42,31,40,54,30,15,35,19,23,22,18]

# lets calculate the standard deviation and mean
mean = 0
for yval in y:
    mean += yval

mean = mean / len(y)
sd = 0
for yval in y:
    sd += (yval - mean) ** 2
    
sd = sd / (len(y)-1)
sd = sd ** 0.5
print(mean)
print(sd)
```
